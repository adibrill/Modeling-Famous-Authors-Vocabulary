# 1 Introduction and Literature Review

## 1.1 Introduction

The proposal that preceded this project explored the lexical changes that famous writers' works had undergone in the last few centuries. 

The ground for the research was the numerous historical life-changing events that had occurred during the last two centuries and the assumption that they had enriched authors' works at the time due to globalization and improved communication [1].

The first step of the research was collecting the artwork of eight different authors from the relevant centuries. Two groups of four authors were chosen from the 16-18th and 19-20th centuries, respectively. They were selected according to the rank of most influential authors of their times. For each author, pieces were selected in the following manner: their first published piece, last published piece, and a piece from the prime of their career. That methodology was selected to examine their language use change over time.  

After cleaning and preparing the data, several methods for measuring lexical diversity were presented during the project proposal. Among those methods, I have selected and calculated TTR, MLTD, HDD, and TF-IDF scores for every literal piece.

Another new indice I calculated was the area under the word density curve.
It was calculated by summing the distances between different appearances of the same word. Those sums were then 
plotted with a histogram, and the area beneath the KDE line represents this indice. The area under this curve will increase in two cases, when there are many different words in the text and when the appearances of these words in the text are distant from one another. That is another way to ensure lexical richness. 

The most definite conclusion of the proposal was that the selected indices were not comparable. 
Some literal pieces received low scores by specific methods and high scores from others, but the methods are unrelated. As a result, a decision was taken to treat lexical proficiency as an unsupervised data field.
Therefore, It was decided to look at the research question from a different perspective. 

The current project will determine if the 19th and 20th centuries innovations had a measurable and distinct impact on the lexical richness among writers of that time, such that it would set them apart from writers of the 16th to the 18th century. In addition, the project will attempt to answer whether or not it is possible to divide writers or pieces into groups by their lexical richness attributes.

The extension of this research will explore an unsupervised machine learning practice to find common attributes of lexical richness between the different pieces of each time. It will involve clustering by K-means and GMM algorithms of all indices calculated in the last part of the project. The results of the project will compare the statistical tests' results of the two periods and the statistical tests' results of the clusterization groups, visualize the results and try to decide if, indeed, there is a connection between the division of the groups, and whether historic events enhanced writers' lexical proficiency [2].

### 1.4.1 Statistical tests for centuries groups 



The project’s first step will perform statistical tests on the different indices [26-27]. 
The tests will check for each indice the distribution and homoscedasticity.

A QQ plot will be used to visualize the normality approximation. To determine whether the data is gaussian, every set of features created will pass the Shapiro-Wilk normality test with a confidence rate of 95% (Reject ‘H0: Data is Gaussian’ when p-value <= 0.05).

Next, The data homogeneity will be tested (Homoscedasticity) and the samples’ distributions equality using Levene’s and Bartlett’s tests.
Levene’s test will be used if the data is not normally distributed. Bartlett’s test will take its place if the data is normally distributed since it is more sensitive than Levene’s test to depart from normality.[13]

Levene’s test is sometimes used as a predeceasing step to comparing means, informing the decision on whether to use a pooled t-test or Welch’s t-test. (However, it was shown that such a two-step procedure might markedly inflate the type 1 error obtained with the t-tests and thus should not be done in the first place.[2] Instead, the choice of pooled or Welch’s test should be made a priori based on the study design. [14])

The next analysis stage will check that the sample’s means are equal. This is essential to determine if they belong to the same population. According to the previous tests’ results, if the data samples are both normally distributed and homogeneous, we will use a t-test. 

Mann-Whitney test will be used if the data samples are not normally distributed but share the same variance.
We should check that the distributions of the two samples are alike even though not gaussian since an assumption for the Mann-Whitney U test is that the distribution shapes of the data being tested are roughly even. 

If the data samples are not normally distributed and do not have equal variance, a Mann-Whitney test will be performed since it is more robust for slightly skewed data samples.

All tests are performed at a
95% confidence (i.e. alpha = 0.05).    

The tests mentioned above are safe to use under the following assumptions:

• Continuous data: All data samples are continuous, which is required to select the tests.

• Number of samples: We have at least 30 samples for each indice, which is sufficient sample size for all tests. 

• Independence: The Mann-Whitney U test assumes samples are independent.

All entries in the data are independent because they were taken from different points in time.
Aside from that, the Mann-Whitney U test assumes that samples are independent, sample groups are not related, but that samples’ distributions are not vastly different [25].

• Distribution shape: a similar distribution between samples is assumed for the Mann-Whitney U test. The t-test assumes normal distribution for the samples. The Shapiro-Wilk test was included to indicate the sample’s distribution, and 
If the samples are not gaussian or not shaped similarly, a matching test or no test will be performed.

• Choice of libraries: The SciPy library is used for Shapiro-Wilk, Levene’s Test, Bartlett’s test, t-test, and Mann-Whitney test. SciPy library performs statistical tests efficiently. It is a robust python library for statistical tests.

### 1.4.2 Dimensionality reduction using Principal Component Analysis (PCA)

Principal component analysis, or PCA, is a statistical procedure that allows to summarize of the information contained in large data tables utilizing a smaller set of “summary indices.” 

PCA is used in exploratory data analysis and for making predictive models. It is commonly used for dimensionality reduction by projecting each data point onto only the first few principal components to obtain lower-dimensional data while preserving as much of the data’s variation as possible.

[10] PCA is an orthogonal linear transformation that transforms the data into a new coordinate system. The data’s most significant variance by some scalar projection lies on the first coordinate (called the first principal component), and the second most significant variance lies on the second coordinate. Any additional coordinates are in descending order of the data variance.

This analysis will use PCA to reduce the different indices to groups of 2 and 3 principle components that can fit into a clusterization model. 

The reason for choosing PCA is that machine learning algorithms converge faster when trained on principal components than the original dataset. Moreover, PCA counteracts the issues of high-dimensional data, and it will allow visualization of the clusterization using import matplotlib and seaborn libraries.

Dimensionality reduction techniques reduce machine learning models’ overfitting and increase the ability of the model to generalize high-dimensional data past the training set examples.

PCA was selected in this case because it helps to maintain the data behavior while eliminating linear dependencies by creating linear combinations of the dependent variables. That way, the data remaining is only constructed of independent variables.
Moreover, visualization of more than three dimensions is hard to read. Through PCA, we can easily use all calculated indices as degrees of freedom and include them in the clusterization process.


### 1.4.3 KMeans clusterization and GMM (Gaussian Mixture Model)

K-means clustering is a vector quantization method that aims to partition n observations into k clusters in which each observation belongs to the cluster with the nearest mean. K-means uses squared Euclidean distances and minimizes within-cluster variances. 
[6] advantages of the k-means algorithm are: Relatively simple to implement, scales to large data sets, 
and guarantees convergence. Easily adapts to new examples and generalizes to clusters of different shapes and sizes, such as elliptical clusters.


In statistics, a mixture model is a probabilistic model representing the presence of subpopulations within a population. The model does not require that an observed data set identify the sub-population to which an individual observation belongs. Formally a mixture model corresponds to the mixture distribution that represents the probability distribution of observations in the overall population. However, while problems associated with "mixture distributions" relate to deriving the properties of the overall population from those of the sub-populations, "mixture models" are used to make statistical inferences about the properties of the sub-populations given only observations on the pooled population, without sub-population identity information.

[5] GMM and K-means are similar in allowing an iterative model refinement process to find the best congestion. However, the K-means algorithm differs in the method used for calculating the Euclidean distance while calculating the distance between the two data items. GMM, on the other hand, uses a probabilistic assignment of data points to clusters.
GMM is based on the normal distributions of its variables. The normal distribution relies on the Gaussian function, widely used in statistics and probability theory. The Gaussian functions appear as the density function of the normal distribution. [29] As a result, the GMM algorithm is based on the Gaussian density function.

Due to the difference between those processes, we would like to present both practices to test whether they will have a common outcome.

Shapiro-Wilk Test

The Shapiro-Wilk test evaluates a data sample and quantifies the likelihood that the data was drawn from a Gaussian distribution, named for Samuel Shapiro and Martin Wilk. In practice, the Shapiro-Wilk test is believed to be a reliable normality test. However, some suggest that the test may be suitable for smaller data samples (e.g., thousands of observations or fewer).

The shapiro() SciPy function will calculate the Shapiro-Wilk on a given dataset. The function returns both the W-statistic calculated by the test and the p-value.

```python
def Shapiro_Wilk_Test(df,col):
    #try:
    # If your p value is less than 0.05, which it is,
    # then you reject the null hypothesis and conclude that your data is nonormal.
        stat, p = shapiro(df[col])
        print('Statistics=%.3f, p=%.3f' % (stat, p))
        # interpret
        alpha = 0.05
        if p > alpha:
            print('Sample looks Gaussian (fail to reject H0)')
            return True,stat,p
        else:
            print('Sample does not look Gaussian (reject H0)')
            return False,stat,p
    #except:
        #print("error")
        #return False,0,0
```

Independent T-test

This is a test for the null hypothesis that two independent samples have identical average (expected) values. This test assumes that the populations have identical variances by default. A paired t-test is used when we are interested in the difference between two variables for the same subject. Often the two variables are separated by time. 

```python
def independent_t_test(df1 , df2):
    try:
        #If the p-value is less than your significance level (e.g., 0.05), you can reject the null hypothesis. 
        #The difference between the two means is statistically significant.
        #Your sample provides strong enough evidence to conclude that the two population means are not equal.
        
        stat,p_value = pystats.ttest_ind(df1, df2,equal_var=False)

        alpha = 0.05
        if p_value > alpha:
            print('Indice is not significantly different between groups.(fail to reject H0)')
            return stat,p_value
        else:
            print('Indice is significantly different between groups.(reject H0)')
            return stat,p_value
    except:
        print("error")
        return 0,0
```

Levene’s test

Levene’s test can be used to check the Homogeneity of variances when the data is not drawn from a normal distribution.

```python
from scipy.stats import levene
def levene_test(df1,df2):
    try:
        # define alpha
        alpha =0.05
        # now we pass the groups and center value from the following
        # ('trimmed mean', 'mean', 'median')
        w_stats, p_value =levene(df1,df2, center ='mean')
        print(w_stats, p_value)

        if p_value > alpha :
            print("The variance between groups is not significantly different. (Fail to reject H0)")
            return True 
        else:
            print("The variance between groups is significantly different. (Reject H0)")
            return False
    except:
        print("error")
        return False
```

Mann Whitney test

The Mann-Whitney U-test tests two independent samples
The Mann Whitney U test, sometimes called the Mann Whitney Wilcoxon Test or the Wilcoxon Rank Sum Test, is used to test whether two samples are likely to derive from the same population

```python
def mannwhitneyu_test(data1,data2):
    try:
        mw_stat, mw_p_value = pystats.mannwhitneyu(data1, data2,alternative="less")
        print('The Man-Whitney U test p-value is', mw_p_value)
        if mw_p_value > 0.05:
        # Do not reject null hypothesis
            print('Indice is not significantly different between groups. \n')
        else:
        # Reject null hypothesis
            print('Indice is significantly different between groups. \n')
        return mw_stat, mw_p_value
    except:
        print("error")
        return 0,0
```

ANOVA

ANOVA checks the impact of one or more factors by comparing the means of different samples. ANOVA is functional in cases where more than two groups must be compared (instead of a t-test).

```python
def ANOVA(df1,df2,df3,df4,df5):
    try:
        # stats f_oneway functions takes the groups as input and returns ANOVA F and p value
        fvalue, pvalue = stats.f_oneway(df1,df2,df3,df4,df5)
        print('The ANOVA test p-value is', pvalue)
        if pvalue > 0.05:
        # Do not reject null hypothesis
            print('Indice is not significantly different between groups. \n')
        else:
        # Reject null hypothesis
            print('Indice is significantly different between groups. \n')
        return fvalue, pvalue
    except:
        print("error")
        return 0,0
```

Kruskal Wallis

The Kruskal-Wallis H test (also called the "one-way ANOVA on ranks") is a rank-based nonparametric test that can be used to determine if there are statistically significant differences between two or more groups of an independent variable.

```python
from scipy import stats
def kruskalwallis(df1,df2,df3,df4,df5):
    try:
        # stats kruskal functions takes the groups as input and returns
        stat, pvalue = stats.kruskal(df1,df2,df3,df4,df5)
        print('The kruskalwallis test p-value is', pvalue)
        if pvalue > 0.05:
        # Do not reject null hypothesis
            print('Indice is not significantly different between groups. \n')
        else:
        # Reject null hypothesis
            print('Indice is significantly different between groups. \n')
        return stat, pvalue
    except:
        print("error")
        return 0,0
```

### 5.1.2 Shapiro-Wilk test

Next, we will perform a Shapiro-Wilk test to ensure Gaussian distribution.
We will divide the data into two groups according to the book's publishing year.
The first group will contain books published before 1800, and the second group will contain books published after 1800.
We will perform the Shapiro-Wilk test on each group separately.
The test results will be kept in different dataframes as a utility for the rest of the analysis.
The QQ-plot will help us assess how far the data points from a Gaussian distribution are.

```python
# keep a version of the dataset where there are only the indices columns
#and the publishing year so that it can be used for centuries classification

indices_df = books_data[["Year Published","MATTR","HDD","Normalized_TF-IDF","Normalized_MLTD",
                    "Unique_Tokens_Ratio","Density_Curve_Area_normalized"]]

# Devide the data into two groups according to publishing year - taken from the original dataset
indices_df_16_18 = indices_df[indices_df["Year Published"] < 1800]
indices_df_19_20 = indices_df[indices_df["Year Published"] > 1800]

# reset indexes
indices_df_16_18 = indices_df_16_18.reset_index(drop=True)
indices_df_19_20 = indices_df_19_20.reset_index(drop=True)

# Remove the year of publishment column
indices_df_16_18 = indices_df_16_18.drop(columns=["Year Published"])
indices_df_19_20 = indices_df_19_20.drop(columns=["Year Published"])

# create a new dataframe for the normality and homogenity tests results
pre_class_stats_16_18 = pd.DataFrame(columns=['Indice', 'Is_gaus','Homogenity','Normal_p'])
pre_class_stats_19_20 = pd.DataFrame(columns=['Indice', 'Is_gaus','Homogenity','Normal_p'])

# Copy the indices into the "Indice" column
pre_class_stats_16_18['Indice'] = indices_df_16_18.columns[0:]
pre_class_stats_19_20['Indice'] = indices_df_19_20.columns[0:]


def SW_Test(IndicesDF,pre_class_DF):
    try:
        # perform the shapiro wilk test on each indice from the original dataset to determine normal distribution
        for row in IndicesDF:
            print(row," :")

            # Perform the test
            is_gaus,stats,p = Shapiro_Wilk_Test(IndicesDF,row)
            print()

            # Keep the result to the results dataframe - for further analysis
            pre_class_DF['Is_gaus'][pre_class_DF['Indice']==row]= is_gaus
            pre_class_DF['Normal_p'][pre_class_DF['Indice']==row]= p

            # Print the qq-plot of the distribution
            _qq_plot(IndicesDF,row)
    except:
        print("error")
    return pre_class_DF


pre_class_stats_16_18 = SW_Test(indices_df_16_18,pre_class_stats_16_18)
pre_class_stats_19_20 = SW_Test(indices_df_19_20,pre_class_stats_19_20)
```

### 5.1.3 Homogeneity of variances test

Next, a homogeneity of variances test must be conducted to attend the t-test assumption.
The results will be kept in the results dataframes created in section 5.1.2.


```python
def gaus_test_and_homoscad_tests(group1,group2,stats_group_1,stats_group_2):

    try:
        # for each indice (column) find out if it has passed the distribution test for both groups
        for col in range(0,len(group1.columns)):
            print(stats_group_1['Indice'][col])

            if stats_group_1['Is_gaus'][col] == True and stats_group_2['Is_gaus'][col] == True:
                # Variable is normally distributed - check homogenity using bartlett_test

                is_homogeneus = bartlett_test(group1[group1.columns[col]],
                             group2[group2.columns[col]])

                stats_group_1['Homogenity'][stats_group_1['Indice']==group1.columns[col]]=is_homogeneus
                stats_group_2['Homogenity'][stats_group_2['Indice']==group2.columns[col]]=is_homogeneus

            else:
                # Variable is not normally distributed - check homogenity using levene_test
                is_homogeneus = levene_test(group1[group1.columns[col]],
                             group2[group2.columns[col]])
                stats_group_1['Homogenity'][stats_group_1['Indice']==group1.columns[col]]=is_homogeneus
                stats_group_2['Homogenity'][stats_group_2['Indice']==group2.columns[col]]=is_homogeneus
            print()
    except:
        print("error")

# Call the test function
gaus_test_and_homoscad_tests(indices_df_16_18,indices_df_19_20,pre_class_stats_16_18,pre_class_stats_19_20)
```

### 6.1.3 Kmeans visualization

In order to visualize the new data using k-means properly, we must first find the optimal clusters division for the existing dataframe. We will use the elbow and silhouette methods to find the optimal k clusters.
The elbow method is an empirical method to find out the best value of k. it picks up the range of values and takes the best value among them. It calculates the sum of the points square and the average distance. When the value of k is 1, the within-cluster sum of the square will be high.

Silhouette analysis can study the separation distance between the resulting clusters. The silhouette plot displays how close each point in one cluster is to points in the neighboring clusters and thus provides a way to visually assess parameters like the number of clusters. Ideally, we will choose the k with the highest silhouette score.

Elbow and Silhouette methods are used to find the optimal number of clusters, but the elbow method might be ambiguous as to the optimal value of k. Silhouette analysis can be used to study the separation distance between the resulting clusters and can be considered a better method than the Elbow method.

```python
from mpl_toolkits.mplot3d import Axes3D
import plotly.graph_objs as go
from sklearn.cluster import KMeans
from plotly import tools
from plotly.subplots import make_subplots
import plotly.offline as py
from sklearn.metrics import silhouette_score

# Keep the results for each K tested
distortions = []
silhouettes = []

# Try K's 1 to 10
K = range(1,10)
for k in K:
    # Perform a kmean model for selected K
    kmeanModel = KMeans(n_clusters=k)
    kmeanModel.fit(principalDf)
    # Keep the inertia value of this K
    distortions.append(kmeanModel.inertia_)

    if k > 1:
        # Calculate the silhouette score 
        silhouette = silhouette_score(principalDf, kmeanModel.labels_) 
        
        # Save to the list
        silhouettes.append(silhouette)
        #print("Silhoutte Score = {:.3f}".format(silhouette))
```

```python
# Plot the inertia value for Ks 1 to 10
plt.figure(figsize=(16,8))
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```

It seems that the best fitting K value is 5. That is due to the decrease in difference size between the values before K=5 and after.

The next step is finding the optimal K clusters.
In the k-Means section, we used the inertia and silhouette scores to select the appropriate number of clusters. The GMM algorithm cannot utilize these metrics because they are unreliable when clusters are not spherical or have different sizes. Instead, we will find the model that minimizes a theoretical information criterion, such as “AIC” or “BIC.”

AIC is the Akaike information criterion, an estimator of prediction error and the relative quality of statistical models for a given data set. AIC estimates the relative amount of information lost by a given model: the less information a model loses, the higher the quality of that model. When a statistical model is used to represent the process that generated the data, the representation will rarely be exact; so some information will be lost by using the model to represent the process.

BIC is the Bayesian information criterion (BIC), or Schwarz information criterion (also SIC, SBC, SBIC) is a criterion for model selection among a finite set of models; models with lower BIC are generally preferred.

The optimal K will be the value that grants the minimal value both for the AIC and the BIC criteria.


```python
from sklearn import mixture

GMMData = principalDf_2d[["principal component 1","sqrt principal component 2"]]

# Save the AIC and BIC values found
Sum_bic = []
Sum_aic = []

# Produce GMM model for k 1 to 10
K = range(1,8)
for k in K:
    gmm = mixture.GaussianMixture(n_components=k)
    gmm = gmm.fit(GMMData)
    Sum_bic.append(gmm.bic(GMMData))
    Sum_aic.append(gmm.aic(GMMData))


# Plot both AIC and BIC values found for each K
x1 = K
y1 = Sum_aic
plt.plot(x1, y1, label = "AIC")
x2 = K
y2 = Sum_bic
plt.plot(x2, y2, label = "BIC")

plt.title("AIC and BIC for different numbers of k", fontsize=14, fontweight='bold')
plt.xlabel("k")
plt.ylabel("Information Criterion")
plt.legend(loc='upper right')
plt.show()
```

Conveniently the optimal K for the GMM model is also 5. This is the K value which produces the minimal value for both criteria (k=2 is only a local minimum for AIC, which is why k=5 is preferable in this scenario).
We will generate the GMM model and use a scatter plot to visualize the results. 

```python
gmm = mixture.GaussianMixture(n_components=5, covariance_type='full').fit(GMMData)

#prediction from GMM
gmm_labels = gmm.predict(GMMData)

# Print results
print("GMM Clusterization Results:")
print(gmm_labels)
```

```python
# Plot the Results
sns.scatterplot(data=GMMData, x="principal component 1", y="sqrt principal component 2", 
                hue=gmm_labels,palette=['green','orange','black','blue','red'],legend='full')
plt.gcf().set_size_inches((10, 10)) 
plt.show()
```

## 7.1 Analysis of classifications

Both models have provided us with a classification result at this stage. It is desired to check the level of agreement between the models. Out of all the literal pieces, which pieces were categorized in the same clusters? 

This is a tricky question since the cluster numbers (0,1,2,3) symbolize different groups in each model. Therefore one cannot simply compare the groups by their class identification. 

The forthcoming code section will carry out the following approach: Every two items will "ask" both models if they belong to the same cluster or not. If both models agree that the items belong to the same cluster or not, the algorithm will count the case as a "True Positive" or "True Negative" outcome. In case the models disagree, The algorithm will disregard the outcome. The Accuracy will be calculated as the times the models responded equally divided by the size of the dataset.

```python
# Create a dataset of the combined results
Results_Comparison = pd.DataFrame(columns=['Kmeans_Class', 'GMM_Class'])
Results_Comparison["Kmeans_Class"] = Kmeans_cluster_books_result["Kmeans_Class"]
Results_Comparison["GMM_Class"] = GMM_cluster_books_result["GMM_Class"]
```

```python
from IPython.display import display, HTML
# Start the Accuracy counter at zero
Models_Combined_Accuracy = 0

# Iterate every two items
for i in range(0,len(Results_Comparison)):
    for j in range(0,len(Results_Comparison)):
        if Results_Comparison["Kmeans_Class"][i] == Results_Comparison["Kmeans_Class"][j]:
            if Results_Comparison["GMM_Class"][i] == Results_Comparison["GMM_Class"][j]:
                #print("yes")
                Models_Combined_Accuracy = Models_Combined_Accuracy + 1;
        else:
            if Results_Comparison["GMM_Class"][i] != Results_Comparison["GMM_Class"][j]:
                Models_Combined_Accuracy = Models_Combined_Accuracy + 1;
                
print("The models had an equal decision: ", Models_Combined_Accuracy, " times")          
print("Models' result equality is: ", 
      Models_Combined_Accuracy/(len(Results_Comparison)*len(Results_Comparison))*100, "% .")

Results_Comparison
```

## 8.3 Evaluation of approach

The project used several innovative practices that had not been used before in the field of lexical richness. 

First, The research incorporated multiple different lexical richness measures (including TF-IDF [38]), while a study by Yanhui Zhang1* and Weiping Wu2 [2] mainly focused on variations of TTR and characteristic constants.

The study used Yule K and Yule I, which are functions of all relative frequencies of all types, where the relative frequency of a type is simply the ratio between the number of occurrences of this type and Tokens.

The method used in this research also measured the distances between the words' occurrences, as density function, and the frequency. The study's method assumes that the distribution of the appearances is even, while the project's method considers that the distribution might change along the text. 

The study used t-test and ANOVA for each measure separately, which obligated the research analysis to reflect on each measure separately. 
Thanks to the PCA method, the results were comprehensive for the entire set of indices.

The study only used KMeans classification for the clustering stage, while this project used both KMeans and GMM to validate the models' outcomes. 

Another study that can be compared to the methods used in this research is "Computational assessment of lexical differences in L1 and L2 writing"[7] by Danielle McNamara. The study used the computational tool Coh-Metrix, measuring cohesion and text difficulty at various language levels.
The study inspected texts' coreferentiality, Latent semantic analysis, Word frequency, and more. These measures were tested individually and then were used for the discriminant function analysis to determine group classification. 
The results of the model accuracy were 79%. However, another model did not cross-examine the results. On top of that, the indices used for the analysis are complicated to measure (such as hypernymy and polysemy indices and causal cohesion).
Another obstacle is using the Discriminant function, which assumes the data is normally distributed.

In this project, the KMeans algorithm is used to validate the results of the GMM classification. Due to KMeans' reliance on euclidean distance, the data distribution was not required to be Gaussian. 
The indices used for classification were relatively simple to measure from the text.
Both models had 75% matching results, which is close to the compared study's results. 

Furthermore, these studies dealt with comparing L1 and L2 writing, which is usually simpler to differentiate, as L1 and L2 writers will vary greatly in terms of lexical depth of knowledge, variation, and sophistication.

For the reasons listed above, I believe the current project shows distinction and novelty in the field of lexical richness study.
